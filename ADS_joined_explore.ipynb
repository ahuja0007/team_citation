```python
import ads
import pandas as pd
import cPickle as pickle
import numpy as np
import matplotlib.pyplot as plt
import math as math
import scipy.stats as stats
%matplotlib inline 
```

```python
# Load in data
filename = 'astro-phstart0max5000'
pkl_file = open('./results_arxiv_api/' + filename + '.pkl','rb')
arXiv_dict = pickle.load(pkl_file)
# Or, equivalently: 
# arXiv_dict = pickle.load(file('./results_arxiv_api/astro-phstart0max1000.pkl'))
```

```python
# def create_df(papers_list):
#     return pd.DataFrame({'Title':[foo.title for foo in papers_list],
#                   'Author':[foo.author for foo in papers_list],
#                   'Citation_count':[foo.citation_count for foo in papers_list],
#                   'Year':[foo.year for foo in papers_list],
#                   'Pub':[foo.pub for foo in papers_list]})

# Parse through comments to extract page and figure numbers.
def search_comments(comments_string,page_numbers=np.NAN,figure_numbers=np.NAN):
    comments = comments_string.split(' ')
    page_number_ind = -1
    figure_number_ind = -1
    for i,com in enumerate(comments):
        if com.find('pages')!=-1:
            page_number_ind = i-1
        if com.find('figures')!=-1:
            figure_number_ind = i-1
    if page_number_ind!=-1:
        if comments[page_number_ind].isdigit():
            page_numbers = int(comments[page_number_ind])
    if figure_number_ind!=-1:
        if comments[figure_number_ind].isdigit():
            figure_numbers = int(comments[figure_number_ind])
        
        
    return [page_numbers,figure_numbers]
```

```python
# Function to generate dataframe aggregating all the data

def generate_df(arXiv_dict):
    
    df = pd.DataFrame()

    for i,key in enumerate(arXiv_dict.keys()):
        pages,figures = search_comments(arXiv_dict[key]['comment'])

        try:
            ads_paper = list(ads.query(title=str(arXiv_dict[key]['title'])))
            if len(ads_paper)==1:  
                new_row = pd.DataFrame({'Arxiv_key': [arXiv_dict.keys()[i]], 'Title':[ads_paper[0].title],
                                            'Author':[ads_paper[0].author],
                                            'Citation_count':[ads_paper[0].citation_count],
                                            'Year':[ads_paper[0].year],'Pub':[ads_paper[0].pub],
                                            'Pages':[pages],'Figures':[figures]})
            #         else:
            #             print i,'key',key,'has multiple query returns'
                df = df.append(new_row)
        except:
            continue
            #    except error:
            #       continue

            # Save dataframe (query takes a long time)
    df.to_csv(filename+'.csv', index = False)

    return
```

```python
# Actually generate dataframe - do not execute if csv file exists already
generate_df(arXiv_dict)
```

```python
# Load data from existing csv
# Modify to load in more data & fuse dataframes!
df = pd.read_csv(filename+'.csv')
```

```python
# Some exploratory data analysis


# Exclude Citation_count > cutoff
cutoff = 150
cit_xarr = np.arange(0,cutoff)
data = np.array(df.Citation_count[df.Citation_count < cutoff])

# Fit Poisson and Gamma distributions
lambda_MLE = np.mean(data)
poisson_MLE = lambda_MLE**cit_xarr * np.exp(-lambda_MLE) / [math.factorial(item) for item in cit_xarr]
print "Poisson lambda_MLE: ", lambda_MLE
shape, loc, scale = stats.gamma.fit(data)
print "Fitted Gamma shape, loc, scale: ", shape, loc, scale

# Plotting

plt.clf()
plt.hist(data, bins=20, normed=True)
plt.plot(cit_xarr,stats.gamma.pdf(cit_xarr, shape, scale=scale), '-r')
plt.title("Citation count distribution")
plt.xlabel("Citation number")
plt.ylabel("Count")
plt.show()

plt.clf()
plt.plot(cit_xarr, poisson_MLE)
plt.plot(cit_xarr,stats.gamma.pdf(cit_xarr, shape, scale=scale), '-r')
plt.xlabel("Citation count")
plt.ylabel("Probability")
plt.title("Fitted probability distributions")
```

```python
# Just some scatterplots

pub_year = [float(item) for item in df.Year.values]
plt.scatter(pub_year, df.Citation_count)
plt.xlabel("Year of publication")
plt.ylabel("Citation count")
plt.title('Publication year and citations')
plt.show()

plt.clf()
plt.scatter(df.Figures, df.Citation_count)
plt.axis([0, 30, 0, 200])
plt.xlabel("Number of figures")
plt.ylabel("Citation count")
plt.title('Figures and citations')
plt.show()

plt.scatter(df.Pages, df.Citation_count)
plt.axis([0, 30, 0, 200])
plt.xlabel("Number of pages")
plt.ylabel("Citation count")
plt.title('Pages and citations')
plt.show()
```

```python
# Boxplots

figs_list = []
for i in np.arange(1,30):
    subdf = df[df.Figures == i]
    figs_list.append(np.array(subdf.Citation_count.values))

pages_list = []
for i in np.arange(1,30):
    subdf = df[df.Pages == i]
    pages_list.append(np.array(subdf.Citation_count.values))
    
box = plt.boxplot(figs_list)
plt.axis([0,30,0,200])
plt.setp(box['whiskers'], color='black') 
plt.setp(box['boxes'], color='black') 
plt.title('Figures and citations')
plt.xlabel("Number of figures")
plt.ylabel("Number of citations")
plt.show()
    
box = plt.boxplot(pages_list)
plt.axis([0,30,0,200])
plt.setp(box['whiskers'], color='black') 
plt.setp(box['boxes'], color='black') 
plt.title('Pages and citations')
plt.xlabel("Number of pages")
plt.ylabel("Number of citations")
plt.show()
```
