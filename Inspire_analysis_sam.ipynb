```python
## load dataframe from the inspire data augmented with additional features
inspire_data=pd.DataFrame.from_csv('INSPIRE_merged_data.csv')
inspire_data.head()
```

We do our last bit of data wrangling, to obtain one final dataset with the features we have constructed and selected throughout our study. We filter the results such that the analyzed papers are all published in journals with median citation count above 5 (Journal quality), and drop the rows with unknown entries. The statistics are presented below.

```python
##count the number of authors per paper
inspire_data['author_counts']=map(lambda x: len(x.strip('[]').split(','))*1.0/2,inspire_data['Author'])

params=['Figures','Pages','Year','Pub_cit_median','Avg_Cens','Avg_Cols','Avg_Pubs','Max_Cens','Max_Cols','Max_Pubs','Min_Cens','Min_Cols','Min_Pubs','author_counts','h_index','m_index','e_index','i10_index','g_index','tori_index','riq_index']

## Take the features we are interested in
numeric_components=inspire_data[params+['Citation_count']]


print "Total rows: ",len(numeric_components)
numeric_components=numeric_components.dropna()
print "Total rows cleaned from NaN values: ",len(numeric_components)
numeric_components['Figures']=numeric_components['Figures']*1.0/numeric_components['Pages']

#clean for journal quality
numeric_components=numeric_components[numeric_components['Pub_cit_median']>5]
target=numeric_components['Citation_count']
print "Total rows filtered by Journal quality: ", len(numeric_components)

#replacing headers with human friendly values
numeric_components.rename(columns={'Figures': 'Figures per page', \
                                   'Pages': 'Total # of pages',\
                                   'Pub_cit_median': 'Journal Quality',\
                                   'Avg_Cens':'Avg centrality',\
                                   'Avg_Cols':'Avg # of collaborators',\
                                   'Avg_Pubs': 'Avg # of pubs',\
                                   'Max_Cens':'Max centrality'\
                                   ,'Max_Cols':'Max # of collaborators'\
                                   ,'Max_Pubs': 'Max # of pubs'
                                   ,'Min_Cens': 'Min centrality'\
                                   ,'Min_Cols': 'Min # of collaborators'
                                   ,'Min_Pubs': 'Min # of pubs'
                                   ,'author_counts':'# of authors'
                                   }, inplace=True)
params=list(numeric_components.columns)
params.remove('Citation_count')
numeric_components.head()
```

###Analysis
We tried a variety of regression methods (linear, polynomial, support vector) on this dataset, however the prediction power was extremely low. Dimentionality reduction did not help this either. Hence, we decided to try classification. For this purpose, we divided the data based on citation counts into three categories:(1) Papers with less than 20 citations, (2) papers with 20-50 citations, and (3) papers with more than 50 citations. The choice of these boundaries was ad hoc. However later analysis also revealed that our data set is well balanced with these boundaries. The distribution of classes is provided below. 

```python
#Helper function to categorize classes based on the number of citations
def class_citations(element):
    if element<21:
        return 1
    elif element <51:
        return 2
    else:
        return 3

#Generate categories
target_classes=np.array(map(class_citations,target))

#Helper function to label the percentages 
def autolabel(rects,freqs):
    # attach some text labels
    for rect,f in zip(rects,freqs):
        width = int(rect.get_width())
        yloc = rect.get_y()+rect.get_height()/2.0

        plt.text(width*0.8, yloc,str(f)+"%", horizontalalignment='left',
            verticalalignment='center', color='white', weight='bold')

#Calculate class abundance and frequency
counts=[]
freqs=[]
for i in range(1,4):
    counts.append(list(target_classes).count(i))
    freqs.append(round(list(target_classes).count(i)*100.0/len(list(target_classes)),2))
    
rects=plt.barh(range(3),counts,align='center',alpha=0.8,color='g');
plt.yticks(range(3),['<=20 citations','21-50 citations', '50< citations' ]);
plt.xlabel('# of datapoints in each category');
plt.title('Distribution of papers across categories');
autolabel(rects,freqs)
```

We repeated our PC analysis on the categorized data. As it is seen in the 2-dimensional example below, there is no obvious trend in how the categories separate. 

```python
#PC analysis
from sklearn.decomposition import PCA

#take the entire dataset
X_train=numeric_components[params]
y_train=target_classes

pca=PCA(n_components=2, copy=True, whiten=False)
X_new=pca.fit_transform(X_train)

plt.figure(figsize=(8, 6))
legend_points=[1,2,3]
category={1:"<21 citations",2:"21-50 citations",3:">50 citations"}

for y in range(len(y_train)):
    colors=['green','cyan','red']
    #label data on based on class and color (slight trick to get the legend)
    if legend_points and y_train[y]==legend_points[0]:
        plt.scatter(X_new[y,0],X_new[y,1],s=20,c=colors[y_train[y]-1],label=category[y_train[y]],alpha=0.6,edgecolors='none')
        legend_points.pop(0)
    else:
        plt.scatter(X_new[y,0],X_new[y,1],s=20,c=colors[y_train[y]-1],alpha=0.6,edgecolors='none')

    
plt.title('Transformation of data onto two principal components')
plt.xlabel('PC1')
plt.ylabel('PC2')

plt.legend(scatterpoints=1)
plt.show()
```

###SVM
Moving on from regression, we tried using SVMs to classify our labeled data. We tried using SVMs on the full feature set, reduced dimensionality using PCA, and reduced feature set (ad hoc selection of features, with some manual optimization). A summary of the results for this analysis is presented below.

```python
from sklearn import svm, grid_search
#run svm analysis on PC transformed data with 10 dimensions
print "SVM with PCA, n_components=10"

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25, random_state=0)

pca=PCA(n_components=10, copy=True, whiten=False)
X_train_new=pca.fit_transform(X_train)
X_test_new=pca.transform(X_test)

svr = svm.SVC()
clf = grid_search.GridSearchCV(svr,{'C':[1, 10, 100]})
clf.fit(X_train_new,y_train)

print "training data score: ", clf.score(X_train_new,y_train)
print "test data score: ", clf.score(X_test_new,y_test)
print "------------"

#SVM without PCA
print "SVM without PCA"
X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25, random_state=0)
svr = svm.SVC()
clf = grid_search.GridSearchCV(svr,{'C':[1, 10,100]})
clf.fit(X_train,y_train)
print "training data score: ", clf.score(X_train,y_train)
print "test data score: ", clf.score(X_test,y_test)
print "------------"
#choose 5 features only 
print "SVM without PCA, 5 features only"
few_params=list(numeric_components.columns)[:5]
X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[few_params], target_classes, test_size=0.25, random_state=0)
svr = svm.SVC()
clf = grid_search.GridSearchCV(svr,{'C':[1, 10,100]})
clf.fit(X_train,y_train)
print "training data score: ", clf.score(X_train,y_train)
print "test data score: ", clf.score(X_test,y_test)
```

As it can be seen on the printout above, the SVM approach is suffering from overfitting for the full feature set, as well as the PC transformed components. This is severe enough to render the SVM predicitons useless (same as guessing at random). Reducing the number of features helps the problem, and we can get slighly better than random performance. This approach out performed the regression methods we had tried before, although the benchmark set by the regression methods was practically no better than random guessing.

Note: We did not use CV for training data scores simply because the performance is high due to overfitting, and using cross-validation would not add any new insight (but slow down the analysis).

###Boosting
Due to the problems that SVMs had with overfitting, we diverted our attention to decision trees and in particular the family of boosting algorithms. The classification results improved significantly. One additional positive of this approach was that the algorithm gave us a description of the features that it used. Below we present the results for two related boosting algorithms: Adaboost and gradient boosting.

```python
#helper function for calculating mean test data performance as well as best features
def mean_test_score_and_feature_importance(classifier,X,Y,num_trials):
    scores=[]
    feature_importances=[]
    for i in range(num_trials):
        X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,Y, test_size=0.25)
        clf=classifier.fit(X_train,y_train)
        scores.append(clf.score(X_test,y_test))
        if feature_importances:
           for i in range(len(feature_importances)):
                feature_importances[i]+=clf.feature_importances_[i]*1.0/num_trials
        else:
            feature_importances=[u*1.0/num_trials for u in clf.feature_importances_]
            
    return scores, feature_importances
```

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25, random_state=0)
ABC=AdaBoostClassifier(n_estimators=50)
scores = cross_validation.cross_val_score(ABC, X_train,y_train, cv=5)
print "Adaboost"
print "CV mean score:", np.mean(scores)
scores, feature_importances_abc=mean_test_score_and_feature_importance(ABC,numeric_components[params],target_classes,20)       
print "test data mean score: " , np.mean(scores)

print "------------"

print "Gradient Boosting" 
X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25,random_state=0)
GBC=GradientBoostingClassifier(loss='deviance', learning_rate=0.1, n_estimators=50, subsample=1.0, min_samples_split=2, min_samples_leaf=1, max_depth=3, init=None, random_state=None, max_features=None, verbose=0, max_leaf_nodes=None, warm_start=False)
scores = cross_validation.cross_val_score(GBC, X_train,y_train, cv=5)
print "CV mean score: ",np.mean(scores)
#get mean test score and feature importances
scores, feature_importances_gbc=mean_test_score_and_feature_importance(GBC,numeric_components[params],target_classes,20)       
print "test data mean score: " , np.mean(scores)
```

To avoid overfitting, we kept our tree-depth and number of trees in the ensemble, and learning rate low (after trying a variety of parameters with gridsearch). 

As it can be seen by the results of the boosting algorithms, the performance on the test data is much better than SVMs(\~0.44). Using this approach, we can predict the correct class almost twice the accuracy of random guessing. The algorithms also provide us with a list of features they have used to make this classification. 

```python
plt.figure(figsize=(12, 6))

plt.subplot('121')

sorted_features2=sorted(zip(feature_importances_abc,params))
importances2,params_sorted2=zip(*sorted_features2)
plt.ylim([-1,len(params)])
plt.barh(range(len(params)),importances2,align='center',alpha=0.6,color='g')
plt.yticks(range(len(params)),params_sorted2,fontsize=12,);
plt.ylabel('feature',fontsize=12);
plt.xlabel('mean importance',fontsize=12);
plt.title('Mean feature importances\n for adaboost classifier');

plt.subplot('122')
sorted_features1=sorted(zip(feature_importances_gbc,params))
importances1,params_sorted1=zip(*sorted_features1)
plt.ylim([-1,len(params)])
plt.barh(range(len(params)),importances1,align='center',alpha=0.6,color='g')
plt.tick_params(axis='y', which='both', labelleft='off', labelright='on')

plt.yticks(range(len(params)),params_sorted1,fontsize=12);
plt.xlabel('mean importance',fontsize=12);
plt.title('Mean feature importances\n for gradient boosting classifier');
```

The two classifier agree on the key features to a great extent. The results also agree with our intuition as well as our exploritory analysis. For instance, we found in our exploraroty analysis that the year of publication did not greatly affect the probability of getting a highly cited paper. The classifiers agree. Like many other studies, we found that the jounral quality is an important factor in the number of citations a paper receives, although we are aware that paper quality and journal quality may be correlated strongly. Below, we have combined the results from the two classifiers and discuss the most important features. 

```python
combined_score=[c+d for c,d in zip(feature_importances_abc,feature_importances_gbc)]
sorted_features3=sorted(zip(combined_score,params))
importances3,params_sorted3=zip(*sorted_features3)
plt.figure(figsize=(12, 6))
p1=plt.barh(range(len(params)),importances1,align='center',color='purple',label='GradientBoost',alpha=0.6)
p2=plt.barh(range(len(params)),importances2,align='center',color='orange',left=importances1,label='Adaboost',alpha=0.6)
plt.ylim([-1,len(params)])
plt.yticks(range(len(params)),params_sorted3,fontsize=12);
plt.ylabel('feature',fontsize=12);
plt.xlabel('importance',fontsize=12);
plt.title('Mean feature importances combined');
plt.legend(loc=4);
plt.show()
```

The combined feature importances indicate that the *e_index, i10_index,* and *h_index* of the first author before the publication and the *length of the paper*, as well as the *quality of the journal* in which it was published, and the  are the top 5 important features for these classifiers to predict the category in which the papers fall. It is useful to look at our analysis on the correlations between these features and the number of citations to understand the full picture. It is also of note that 3 out of the top 5 features are related to the author, rather than properties that are captured by the paper data alone.

###Combining our classifiers

We also investigated whether the three classifiers (SVM,Adaboost, Gradient boosting) can be improved together by implementing a voting mechanism. Namely, if there is a way that two classifiers can correct the mistakes of the other one. This is dependent on the structure and composition of the data, and originally when we tried this mechanism, the voting would actually improve classification. However, upon adding more data (and features) to our dataset, the voting lost it's advantage to the best classifier. We tried different voting schemes, but none improved the outcome. The results are presented below. 

*Note*: we are aware that we could use Adaboost on this layer again to optimize weights for the voting scheme. However, the runtime was very long and the code was complex to implement, hence we abandoned this additional attempt. 

```python
#Compute the most common vote 
def most_common(lst):
    if len(set(lst))==3:#if all votes are different pick the best classifier (always GBC in this case)
        return lst[2]
    return max(set(lst), key=lst.count)

svmscore=[]
gbcscore=[]
abcscore=[]
combinedscore=[]

# try 30 trials 
for i in range(30):
    # pick the same random state for all classifiers (different for each trial)
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[few_params], target_classes, test_size=0.25,random_state=i)
    #make a combined dataframe for all votes, store the real answer
    combined_predictions=pd.DataFrame(columns=['answer','SVM','GBC','ABC'])
    combined_predictions['answer']=y_test

    svr = svm.SVC()
    clf=svr.fit(X_train,y_train)
    combined_predictions['SVM']=map(clf.predict,X_test)
    svmscore.append(clf.score(X_test,y_test))
    
    X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25,random_state=i)

    
    GBC=GradientBoostingClassifier(loss='deviance', learning_rate=0.1, n_estimators=50, subsample=1.0, min_samples_split=2, min_samples_leaf=1, max_depth=3, init=None, random_state=None, max_features=None, verbose=0, max_leaf_nodes=None, warm_start=False)
    clfgbc=GBC.fit(X_train,y_train)
    combined_predictions['GBC']=map(clfgbc.predict,X_test)
    gbcscore.append(clfgbc.score(X_test,y_test))

    ABC=AdaBoostClassifier(n_estimators=50)
    clfabc=ABC.fit(X_train,y_train)
    
    combined_predictions['ABC']=map(clfabc.predict,X_test)
    abcscore.append(clfabc.score(X_test,y_test))


    combined_predictions['voted']=[[a[0],b[0],c[0]] for a,b,c in zip(combined_predictions['SVM'],combined_predictions['GBC'],combined_predictions['ABC'])]
    combined_predictions['predictions']=map(most_common,combined_predictions['voted'])
    
    
    combinedscore.append(len(combined_predictions[combined_predictions['answer']==combined_predictions['predictions']])*1.0/len(combined_predictions))
```

The relative performance of the classifiers for each trial is depicted below

```python
plt.figure(figsize=(8, 6))

plt.plot(range(30),svmscore,label='SVN',linewidth=2.0)
plt.plot(range(30),gbcscore,label='gradientboost',linewidth=2.0)
plt.plot(range(30),abcscore,label='adaboost',linewidth=2.0)
plt.plot(range(30),combinedscore,label='combined voting',linewidth=2.0)
plt.legend(bbox_to_anchor=(1.4, 0.7))

plt.xlabel('Trail #')
plt.ylabel('Accuracy')
plt.title('Relative performance for different classifiers')
plt.show()
```

The gradient boost classifier always outperforms every other classifier we used, even if we try to combine all the classifiers and use naive voting to classify our points.
