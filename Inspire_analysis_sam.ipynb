```python
%matplotlib inline 

import numpy as np
import pickle
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import random
import datetime as dt
import json
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn import cross_validation
from itertools import *
import random
```

```python
inspire_data=pd.DataFrame.from_csv('INSPIRE_merged_data.csv')
inspire_data.head()
```

```python
inspire_data['author_counts']=map(lambda x: len(x.strip('[]').split(','))*1.0/2,inspire_data['Author'])
numeric_components=inspire_data[['Figures','Pages','Year','Pub_cit_median','Avg_Cens','Avg_Cols','Avg_Pubs','Max_Cens','Max_Cols','Max_Pubs','Min_Cens','Min_Cols','Min_Pubs','author_counts','Citation_count']]
print len(numeric_components)
numeric_components=numeric_components.dropna()
print len(numeric_components)
numeric_components['Figures']=numeric_components['Figures']*1.0/numeric_components['Pages']
numeric_components=numeric_components[numeric_components['Pub_cit_median']>5]
target=numeric_components['Citation_count']
print len(numeric_components)
numeric_components.head()
```

```python
def class_citations(element):
    if element<21:
        return 1
    elif element <51:
        return 2
    else:
        return 3
```

```python
#PC analysis
from sklearn.decomposition import PCA
params=['Figures','Pages','Year','Pub_cit_median','Avg_Cens','Avg_Cols','Avg_Pubs','Max_Cens','Max_Cols','Max_Pubs','Min_Cens','Min_Cols','Min_Pubs','author_counts']
target_classes=np.array(map(class_citations,target))

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.2, random_state=0)


pca=PCA(n_components=2, copy=True, whiten=False)

X_new=pca.fit_transform(X_train)
plt.scatter(X_new[:,0],X_new[:,1],s=20, c=y_train)
plt.colorbar()
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.show()
```

```python
print "class 1",list(target_classes).count(1)*1.0/len(target_classes)
print "class 2",list(target_classes).count(2)*1.0/len(target_classes)
print "class 3",list(target_classes).count(3)*1.0/len(target_classes)
```

```python
#helper function to get all subsets
def powerset(iterable):
    s = list(iterable)
    return chain.from_iterable(combinations(s, r) for r in range(len(s)+1))
```

```python
from sklearn import svm, grid_search

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.2, random_state=0)

pca=PCA(n_components=5, copy=True, whiten=False)
X_train_new=pca.fit_transform(X_train)
X_test_new=pca.transform(X_test)

svr = svm.SVC()
clf = grid_search.GridSearchCV(svr,{'C':[1, 10, 100]})
clf.fit(X_train_new,y_train)


print "training data score ", clf.score(X_train_new,y_train)
print "test data score ", clf.score(X_test_new,y_test)
    
```

```python
from sklearn.ensemble import GradientBoostingClassifier

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25,random_state=0)

GBC=GradientBoostingClassifier(loss='deviance', learning_rate=0.1, n_estimators=50, subsample=1.0, min_samples_split=2, min_samples_leaf=1, max_depth=3, init=None, random_state=None, max_features=None, verbose=0, max_leaf_nodes=None, warm_start=False)
scores = cross_validation.cross_val_score(GBC, X_train,y_train, cv=5)

print np.mean(scores)
clfgbc=GBC.fit(X_train,y_train)
print clfgbc.score(X_test,y_test)
```

```python
sorted_features1=sorted(zip(clfgbc.feature_importances_,params))
importances1,params_sorted1=zip(*sorted_features1)
plt.barh(range(14),importances1,align='center')
plt.yticks(range(14),params_sorted1);
plt.ylabel('feature');
plt.xlabel('importance');
```

```python
from sklearn.ensemble import AdaBoostClassifier

X_train, X_test, y_train, y_test = cross_validation.train_test_split(numeric_components[params], target_classes, test_size=0.25, random_state=0)

ABC=AdaBoostClassifier(n_estimators=50)
scores = cross_validation.cross_val_score(ABC, X_train,y_train, cv=5)

print np.mean(scores)
clfabc=ABC.fit(X_train,y_train)
print clfabc.score(X_test,y_test)
```

```python
sorted_features2=sorted(zip(clfabc.feature_importances_,params))
importances2,params_sorted2=zip(*sorted_features2)
plt.barh(range(14),importances2,align='center')
plt.yticks(range(14),params_sorted2);
plt.ylabel('feature');
plt.xlabel('importance');
```
