```python
#Testing 
%matplotlib inline 

import numpy as np
import pickle
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import random
import datetime as dt

filenames=glob('results_arxiv_api/*')

all_dicts={}
for filename in filenames:
    pkl_file = open(filename, 'rb')
    recovered_pickle=pickle.load(pkl_file)
    all_dicts=dict(all_dicts.items()+recovered_pickle.items())

```

```python
"""Helper function to get a random element from the dictionary"""
def get_sample(data):
    return data[random.choice(data.keys())]

#plot the author counts
def plot_author_counts(data):
    author_counts=[]
    for key in data:
        author_counts.append(len(data[key]['authors']))
    plt.hist(author_counts,bins=100);
    plt.xlim(1,20)
    plt.yscale('log')
    plt.xlabel('number of authors')
    plt.ylabel('number of papers')
    plt.title('Number of authors for each paper')
    plt.show()

plot_author_counts(all_dicts)    
```

```python
#print get_sample(all_dicts)

def plot_author_counts_per_year(data):
    years_covered=[]
    for key in data:
        if int(data[key]['published'][:4]) not in years_covered:
            years_covered.append(int(data[key]['published'][:4]))
    years_covered.sort()
    single_authors=[]
    for year in years_covered:
        year_author_count=[]
        a=0.001
        c=[0,0]
        for key in data:
           if int(data[key]['published'][:4])==year:
              year_author_count.append(len(data[key]['authors']))
              if len(data[key]['authors'])>2:
                  c[1]+=1
              else:
                  c[0]+=1
              a+=1
        single_authors.append([z*1.0/a for z in c])    
    plt.plot(years_covered,[s[0] for s in single_authors],'*-',label="1 or 2 authors")
    plt.plot(years_covered,[s[1] for s in single_authors],'*-',label="more than 2 authors")

    plt.legend()
    plt.ylabel('#of authors on the paper (normalized)')
    plt.xlabel('year')
    plt.show()
 
plot_author_counts_per_year(all_dicts)
```

```python
def num_papers_per_authors(data):
    authors={}
    for key in data:
        for author in data[key]['authors']:
            if author in authors:
                authors[author]+=1
            else:
                authors[author]=1
    plt.hist(authors.values(),bins=75)
    plt.xlim(1,25)
    plt.yscale('log')
    plt.ylabel('authors')
    plt.xlabel('number of papers')
    plt.show()
    
num_papers_per_authors(all_dicts)
```

```python
def plot_title_analysis_per_year(data):
    years_covered=[]
    for key in data:
        if int(data[key]['published'][:4]) not in years_covered:
            years_covered.append(int(data[key]['published'][:4]))
    years_covered.sort()
    title_lengths=[]
    title_std=[]
    years=[]
    for year in years_covered:
        year_title_lengths=[]
        for key in data:
            if int(data[key]['published'][:4])== year:
                year_title_lengths.append(len(data[key]['title']))
        years.append(year_title_lengths)
        title_lengths.append(np.mean(year_title_lengths))
        title_std.append(np.std(year_title_lengths))
    #plt.plot(years_covered,title_lengths,'*-')
    #plt.errorbar(years_covered, title_lengths, title_std, linestyle='--', marker='*')
    for year,num in zip(years,years_covered):

        plt.hist(year,histtype='step',label=str(num),bins=50)
    plt.legend()
    plt.ylabel('Frequency')
    plt.xlabel('length of title')
    plt.show()
plot_title_analysis_per_year(all_dicts)
```
