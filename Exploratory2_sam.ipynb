```python
%matplotlib inline 

import numpy as np
import pickle
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import random
import datetime as dt
import json

included_years=[u'2002', u'2003', u'2000', u'2001', u'2006', u'2007', u'2004', u'2005', u'2008', u'2009']
```

```python
def read_line_by_line(f):
    for line in f:
        yield line
        
        
def get_sample():
    f=open('hep_records.json','r')
    for line in f:
        response=json.loads(line)
        return response
```

```python
def citation_count_by_year(include_years):
    f=open('hep_records.json','r')
    years={}
    for line in f:
        response=json.loads(line)
        try:
            if response['creation_date'][:4] not in included_years:
                continue
            elif int(response['creation_date'][:4]) in years:
                years[int(response['creation_date'][:4])].append(len(response['citations']))
            else:
                years[int(response['creation_date'][:4])]=[len(response['citations'])]
        except:
            #print response
            continue
    return years

years=citation_count_by_year(included_years)
```

```python
number_per_year=[len(years[a]) for a in years]

plt.bar(range(len(years.keys())),number_per_year)
plt.xticks(range(len(years.keys())),years.keys(),rotation=45);
plt.xlabel('year')
plt.ylabel('number of papers')
plt.title('INSPIRE data number of papers per year\n (truncated for 2000-2009)')
plt.show()
```

```python
for year in years:
    plt.hist(years[year],histtype='step',label=str(year),bins=np.arange(0,10000,10))
    plt.yscale('log')
    plt.xscale('log')
plt.legend()
plt.ylabel('log scale frequency')
plt.xlabel('log number of citations')
plt.show()
```

####Insight
the distribution of citation counts stays roughly the same over the years. There is more noise in the larger citation counts, presumably because famous papers appear sporadically? 

```python
def author_count_by_year(include_years):
    f=open('hep_records.json','r')
    years={}
    for line in f:
        response=json.loads(line)
     
        try:
            if response['creation_date'][:4] not in included_years:
                continue
            elif int(response['creation_date'][:4]) in years:
                years[int(response['creation_date'][:4])].append(len(response['co-authors'])+1)
            else:
                years[int(response['creation_date'][:4])]=[len(response['co-authors'])+1]
        except:
            #print response
            continue
    return years

author_counts=author_count_by_year(included_years)
```

```python

plt.boxplot([author_counts[year] for year in author_counts] )
plt.ylabel('number of authors')
plt.xlabel('years')

plt.xticks(range(1,11),author_counts.keys())
plt.yscale('log') 
plt.show()
```

```python
def citation_author_count_by_year(include_years):
    f=open('hep_records.json','r')
    citation_author_counts=[]
    for line in f:
        response=json.loads(line)
        try:
            if response['creation_date'][:4] in included_years:
                citation_author_counts.append((len(response['co-authors'])+1,len(response['citations'])))
        except Exception as e:
            #print response
            #print e
            continue
    return citation_author_counts

cit_author_counts=citation_author_count_by_year(included_years)
```

```python
authors,cit=zip(*cit_author_counts)
citations_per_number_author={}

log_scale=[(0,1)]+[(5**n,5**(n+1)) for n in range(5)]

for key in log_scale:
    citations_per_number_author[key]=[]

for a,c in cit_author_counts:
    for key in citations_per_number_author:
        if a> key[0] and a<=key[1]:
            citations_per_number_author[key].append(c)

to_boxplot=[]
for key in sorted(citations_per_number_author.keys()):
    to_boxplot.append(citations_per_number_author[key])
    
plt.figsize=(20,8)
plt.boxplot(to_boxplot);
#plt.scatter(authors,cit);
plt.yscale('log');
plt.xticks(range(1,8),["<= "+str(i[1]) for i in sorted(citations_per_number_author.keys())],rotation=45);
#plt.xlim(0,500)
#plt.ylim(0,6000)
plt.xlabel('number of authors');
plt.ylabel('(log) citation numbers');
plt.title('Number of citations in relation \n to the number of authors on the paper');

```

```python
def ref_citation_count(include_years):
    f=open('hep_records.json','r')
    citation_author_counts=[]
    for line in f:
        response=json.loads(line)
        try:
            if response['creation_date'][:4] in included_years:
                citation_author_counts.append((len(response['references']),len(response['citations'])))
        except Exception as e:
            #print response
            #print e
            continue
    return citation_author_counts

ref_cit=ref_citation_count(included_years)
```
