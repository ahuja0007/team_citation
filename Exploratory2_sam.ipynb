```python
%matplotlib inline 

import numpy as np
import pickle
import pandas as pd
from glob import glob
import matplotlib.pyplot as plt
import random
import datetime as dt
import json
```

```python
def read_line_by_line(f):
    for line in f:
        yield line
```

```python
def author_count_by_year(include_years):
    f=open('hep_records.json','r')
    years={}
    for line in f:
        response=json.loads(line)
        try:
            if response['creation_date'][:4] not in included_years:
                continue
            elif int(response['creation_date'][:4]) in years:
                years[int(response['creation_date'][:4])].append(len(response['citations']))
            else:
                years[int(response['creation_date'][:4])]=[len(response['citations'])]
        except:
            #print response
            continue
    return years

included_years=[u'2002', u'2003', u'2000', u'2001', u'2006', u'2007', u'2004', u'2005', u'2008', u'2009']
years=author_count_by_year(included_years)
```

```python
number_per_year=[len(years[a]) for a in years]

plt.bar(range(len(years.keys())),number_per_year)
plt.xticks(range(len(years.keys())),years.keys(),rotation=45);
plt.xlabel('year')
plt.ylabel('number of papers')
plt.title('INSPIRE data number of papers per year\n (truncated for 2000-2009)')
plt.show()
```

```python
for year in years:
    plt.hist(years[year],histtype='step',label=str(year),bins=np.arange(0,10000,10))
    plt.yscale('log')
    plt.xscale('log')
plt.legend()
plt.ylabel('log scale frequency')
plt.xlabel('log number of citations')
plt.show()
```

####Insight
the distribution of citation counts stays roughly the same over the years. There is more noise in the larger citation counts, presumably because famous papers appear sporadically?
